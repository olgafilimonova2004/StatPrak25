## Problem 1

Let $f(x) = 2\sin x - \cos{2x}$. Write a function `beauty_derivative_plot` that takes a point $x$ and plots the function $f(x)$ and its derivative $f'(x)$ in the range $x \in [-2\pi, 2\pi]$. The function should also plot the tangent line to the function at the point $x$.

Example of output for $x = -3, -2, 1.7$:

```python
beauty_derivative_plot(-3)
```

```python
beauty_derivative_plot(-2)
```
<img src="https://i.ibb.co/F4zDq8m0/tangent-line-2-0.png" width="600"/>

```python
beauty_derivative_plot(1.7)
```
<img src="https://i.ibb.co/TBmLXg2h/tangent-line-1-7.png" width="600"/>

````python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2 * np.sin(x) - np.cos(2 * x)

def df(x):
    return 2 * np.cos(x) + 2 * np.sin(2 * x)

def beauty_derivative_plot(a):
    x = np.linspace(-2*np.pi, 2*np.pi, 100)
    y = f(x)
    dy = df(x)
    fa = f(a)
    dfa = df(a)
    tangent_x = np.linspace(a - 1, a + 1, 100)
    tangent_y = fa + dfa * (tangent_x - a)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'b', label=r'$f(x) = 2\sin x - \cos 2x$')
    plt.plot(x, dy, 'r--', label=r"$f'(x) = 2\cos x + 2\sin 2x$")
    plt.plot(tangent_x, tangent_y, 'g', linewidth=2, label=f'Tangent at x = {a:.2f}')

    # draw the point
    plt.scatter([a], [fa], color='k', zorder=3)
    plt.vlines(a, 0, fa, linestyle='dashed', color='purple')
    plt.hlines(fa, 0, a, linestyle='dashed', color='purple')
    plt.annotate(f"f'({a:.2f}) = {dfa:.2f}",
             xy=(a, fa),
             xytext=(a - 0.5, fa - 1),
             arrowprops=dict(facecolor='black', arrowstyle='->', connectionstyle="arc3,rad=0.2"),
             fontsize=10, ha='center')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Function, Derivative, and Tangent Line')
    plt.legend()
    plt.grid(True)

    plt.show()
beauty_derivative_plot(1.71)
beauty_derivative_plot(-3)
````

Find which year each species had the largest population. Mark these years in the plot of populations.

````python
max_l=np.max(df['lynx'])
max_c=np.max(df['carrot'])
max_h=np.max(df['hare'])

max_hare_year=df['year'][np.argmax(df['hare'])]
max_lynx_year=df['year'][np.argmax(df['lynx'])]
max_carrot_year=df['year'][np.argmax(df['carrot'])]

plt.figure(figsize=(10, 6))

plt.plot(df['year'], df['hare'], label='Hare', marker='o')
plt.plot(df['year'], df['carrot'], label='Carrot', marker='o')
plt.plot(df['year'], df['lynx'], label='Lynx', marker='o')

plt.xlabel('Year')
plt.ylabel('Population size')
plt.title('Population sizes of hare, lynx, and carrot')

plt.annotate(f'Max hare: {max_hare_year.astype(int)}',
             xy=(max_hare_year, max_h),
             xytext=(max_hare_year, max_h + 10))
plt.annotate(f'Max lynx: {max_lynx_year.astype(int)}',
             xy=(max_lynx_year, max_l),
             xytext=(max_lynx_year, max_l + 10))
plt.annotate(f'Max carrot: {max_carrot_year.astype(int)}',
             xy=(max_carrot_year, max_c),
             xytext=(max_carrot_year, max_c + 10))

plt.legend()
plt.grid(True)

plt.show()
````

- Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

````python
hare_change = np.diff(df['hare'])
lynx_change = np.diff(df['lynx'])
year_change = df['year'][1:]

plt.figure(figsize=(10, 6))

plt.plot(year_change, hare_change, label='Hare population change', marker='o')
plt.plot(year_change, lynx_change, label='Lynx population change', marker='o')

plt.xlabel('Year')
plt.ylabel('Population change')
plt.title('Change of hare and lynx populations over time')

plt.legend()
plt.grid(True)
plt.show()
corr= np.corrcoef(df['hare'], df['lynx'])[0, 1]
print(f'Corralation coefficient: {corr}')
````
