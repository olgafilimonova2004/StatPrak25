Посмотрим, в какие дни недели перелетов больше. Для каждого дня рассмторим долю от всех полетов (в процентах).

```python
import matplotlib.pyplot as plt

days_count=flights_df['DAY_OF_WEEK'].value_counts().sort_index()
plt.bar(days_count.index, days_count.values*100/days_count.values.sum(),
    color='skyblue',
    edgecolor='black',
    linewidth=1.,
    alpha=0.8,
    width=0.7
)
plt.xticks(days_count.index,['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

plt.xlabel('Day of Week')
plt.ylabel('Percentage of Flights')
plt.title('Percentage of Flights by Day of Week')
plt.show()
```

Чаще всего полеты в рабочие дни, реже всего - по выходным, меньше всего в субботу.

Выделим топ-7 авиакомпаний с наибольшим количеством задержек по времени вылета. Указана доля в процентах.

```python
delayed_df=flights_df[flights_df['DEPARTURE_DELAY']>0].reset_index(drop=True)
airline_delay_counts=delayed_df['AIRLINE'].value_counts()
airline_delay_ratio=airline_delay_counts/airline_delay_counts.sum()
airline_delay_ratio=airline_delay_ratio[:7]

plt.bar(airline_delay_ratio.index,airline_delay_ratio*100,
    color='skyblue',
    edgecolor='black',
    linewidth=1.,
    alpha=0.8,
    width=0.7
)
plt.xlabel('Airline')
plt.ylabel('Percentage of Delayed Flights')
plt.title('Top 7 Airlines with the Most Flight Departure/Arrival Delays')
plt.show()

airlines_df[airlines_df['IATA_CODE'].isin(airline_delay_ratio.index)].reset_index(drop=True)
```

В каких аэропортах происходит задержка из-за погоды?

```python
w_df=delayed_df[delayed_df['WEATHER_DELAY'].notna() & delayed_df['WEATHER_DELAY']>0]
w_df=w_df.sort_values(by='WEATHER_DELAY', ascending=False)
thresh=w_df['WEATHER_DELAY'].quantile(0.75)
print(thresh)
w_df=w_df[w_df['WEATHER_DELAY']>thresh]
weather_cond=w_df['ORIGIN_AIRPORT'].value_counts()[:20]
weather_cond
```

Выбрали топ-20 аэропортов, в которых задержки вылета происходят чаще всего и составляют более 57 минут (квантиль 0.75)

```python
from mpl_toolkits.basemap import Basemap
```

```python
all_airports_d=airports_df[airports_df['IATA_CODE'].isin(delayed_df['ORIGIN_AIRPORT'].unique())]
w_cond=airports_df[airports_df['IATA_CODE'].isin(weather_cond.index)]

fig = plt.figure(figsize=(15, 10))
land = Basemap(projection='lcc', resolution=None,
               width=15E6, height=8E6,
               lat_0=45, lon_0=-100,)
land.etopo(scale=0.5, alpha=0.9)

airports_xy=np.array(land(all_airports_d['LONGITUDE'], all_airports_d['LATITUDE'])).T
weather_xy=np.array(land(w_cond['LONGITUDE'], w_cond['LATITUDE'])).T

for a in airports_xy:
    plt.plot(a[0], a[1], marker='o',color='white', markersize=5)
for b in weather_xy:
    plt.plot(b[0], b[1], marker='o',color='blue', markersize=5)
plt.show()
```

Белым отмечены аэропорты, где когда-либо были небольшие задержки с небольшой частотой, синим - аэропорты с частыми задержками более 57 мин. Возможно, эти аэропорты следует разгрузить в самые проблемные времена года - лето и зиму, или готовиться к неприятным погодным условиям заранее (например, уделить больше внимания устранению обледенения самолетов). Заметим также, что большинство синих точек находятся на востоке и юго-востоке страны.
